Q2: Differentiate between one-tailed and two-tailed tests.
ans)

One-tailed and two-tailed tests are types of hypothesis tests used to determine if there is a significant difference between groups.

1. One-Tailed Test
    1. A one-tailed test (or one-sided test) is used when the research hypothesis predicts the direction of the effect.
    2. It tests for the possibility of the relationship in one specific direction.
    3. Example - Tests if a sample mean is significantly greater than or less than a population mean.
    4. Predicts the direction of the effect.
    5. More powerful for detecting an effect in the specified direction
2. Two-Tailed Test
    1. A two-tailed test (or two-sided test) is used when the research hypothesis does not predict the direction of the effect. 
    2. It tests for the possibility of the relationship in both directions
    3. Example - Tests if a sample mean is significantly different from a population mean, in either direction (greater than or less than).
    4. Does not predict the direction of the effect.
    5. Tests for differences in both directions.
    

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for 
each type of error.
Ans)

In hypothesis testing, Type I and Type II errors refer to the potential mistakes we can make when making decisions about the null hypothesis (H0).

1. Type I
    1. A Type I error occurs when we reject the null hypothesis (H0) when it is actually true.
    2. The probability of committing a Type I error is denoted by α (alpha), which is the significance level set by the researcher. In other words there is a 5% chance of rejecting the null hypothesis when it is true.
    
    Example Scenario:
        A drug company tests a new medication to see if it lowers blood pressure more effectively than a placebo. The null hypothesis (H0) is that the new medication has no effect on blood pressure. A Type I error would occur if the test results lead the researchers to conclude that the new medication is effective when, in fact, it is not.
    
2. Type II Error
    1. A Type II error occurs when we fail to reject the null hypothesis (H0) when it is actually false
    2. The probability of committing a Type II error is denoted by β (beta).Researchers aim to maximize power to reduce the likelihood of a Type II error.
    
    Example Scenario:
    A manufacturing company tests whether a new production process reduces defects compared to the current process. The null hypothesis (H0) is that the new process does not reduce defects. A Type II error would occur if the test results lead the researchers to conclude that the new process does not reduce defects when, in fact, it does
    
    

Q4:  Explain Bayes's theorem with an example.
Ans)
Bayes's theorem is a fundamental concept in probability theory that describes the probability of an event based on prior knowledge of conditions that might be related to the event.It allows us to update our beliefs about the likelihood of an event given new evidence.

Example Scenario
Suppose a certain disease affects 1% of the population. There is a test for the disease that is 99% accurate.
This means
    1. If a person has the disease, the test correctly identifies it 99% of the time (true positive rate).
    2. If a person does not have the disease, the test correctly identifies this 99% of the time (true negative rate).
    let's find the probability that a person actually has the disease given that they tested positive.
    
    P(D)=0.01: The prior probability of having the disease (1% of the population).

    P(-D)=0.99: The prior probability of not having the disease (99% of the population).

    P(T|D)=0.99: The probability of testing positive given that the person has the disease (true positive rate).

    P(T∣-D)=0.01: The probability of testing positive given that the person does not have the disease (false positive rate).
    P(T)=P(T|D)⋅P(D)+P(T|D)⋅P(-D) = 0.0198
    P(D∣T)= P(T|D)⋅P(D)/P(T) =  0.5
    
    Conclusion : 
    Despite the test being 99% accurate, if a person tests positive, there is only a 50% chance that they actually have the disease

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Ans)
A confidence interval is a range of values, derived from sample statistics, that is likely to contain the value of an unknown population parameter. The interval has an associated confidence level that quantifies the level of confidence that the parameter lies within the interval.

Confidence Interval=Point Estimate +/- Margin of Error

Example Calculation:
Let's calculate a 95% confidence interval for a sample mean.
Assume below data
    Sample mean (xˉ)  = 50
    Sample standard deviation(s) = 10
    Sample size (n) = 25
    
    1. Determine the z-score for a 95% confidence level:
       For a 95% confidence level, the z-score (critical value) is approximately 1.96.
    2. Calculate the standard error (SE):
            SE = s/√n = 10/√25 = 2
    3. Calculate the margin of error (ME):
            ME = z×SE = 1.96×2 = 3.92
    4. Calculate the confidence interval:
            CI= x bar ± ME = 50±3.92
            CI=(50−3.92, 50+3.92)=(46.08,53.92)
    Conclusion - With 95% confidence, we can say that the true population mean lies between 46.08 and 53.92

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? 
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
Ans)

The margin of error (ME) in a confidence interval quantifies the range within which the true population parameter is expected to lie, given a certain level of confidence.It represents the maximum expected difference between the true population parameter and a sample estimate.

How Sample Size Affects the Margin of Error:
1. The sample size (n) has an inverse relationship with the margin of error. 
2. As the sample size increases, the standard error decreases, leading to a smaller margin of error
3. A larger sample size reduces the variability of the sample mean, providing a more precise estimate of the population parameter.

Example Scenario:
Problem Statement - Estimating the Average Height of Adult Males in a City
Case-1: Sample size is 30.
    Sample size (n) = 30
    Sample mean height (xˉ) = 175 cm
    Sample standard deviation (s) = 10 cm
    Confidence level = 95% (z-score = 1.96)
    
    Calculating the margine of error:
    1. Standard Error (SE): 10/√30 = 1.825
    2. Margin of Error (ME): 1.96×1.825 = 3.577
    3. Confidence Interval(CI):175±3.577 = (171.423,178.577)
Case-2: Sample size is 100
     Sample size (n) = 100
     
     Calculating the margine of error:
    1. Standard Error (SE): 10/√100 = 1
    2. Margin of Error (ME):1.96×1 = 1.96
    3. Confidence Interval(CI):175±1.96=(173.04,176.96)
    
    Conclusion: 
    By increasing the sample size from 30 to 100, the margin of error decreases from 3.577 to 1.96. This results in a narrower confidence interval, indicating a more precise estimate of the average height of adult males in the city.
