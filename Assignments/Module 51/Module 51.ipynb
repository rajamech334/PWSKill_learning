Q2) What is supervised learning? List some examples if supervised learning?

Ans)

Supervised Learning is a type of Machine Learning where the model is trained on labeled data. This means that each training example is paired with an output label, and the model learns to predict the output from the input data. The goal is to make accurate predictions or classifications based on the patterns learned from the labeled examples.

Examples of Supervised Learning:

    1. Classification
        1. Email Spam Detection: Classifying emails as "spam" or "not spam" based on features like subject line, sender, and content
        2. Sentiment Analysis : Determining if a piece of text (e.g., a product review) has a positive, negative, or neutral sentiment.
        
        
    2. Regression
        1. House Price Prediction: Predicting the price of a house based on features such as location, size, and number of bedrooms.
        2. Stock Price Prediction: Estimating future stock prices based on historical data and market indicators.
        
        

Q3) What is unsupervised learning? List some examples if unsupervised learning?

#Ans)

Unsupervised Learning is a type of Machine Learning where the model is trained on data without explicit labels or outcomes. Instead of predicting predefined labels or values, the model tries to uncover hidden patterns, relationships, or structures within the data.

Examples of Unsupervised Learning:

1. Clustering:
    
    1. Customer Segmentation: Grouping customers based on purchasing behavior to tailor marketing strategies.
    2. Document Clustering: Organizing a collection of documents into topics or categories based on content similarity.
    3. Image Segmentation: Grouping pixels in an image into regions with similar colors or textures for further analysis.
    
2. Dimensionality Reduction
    1. Principal Component Analysis (PCA): Reducing the number of features in a dataset while preserving variance, often used for visualization or speeding up other algorithms.
    2. t-Distributed Stochastic Neighbor Embedding (t-SNE): Visualizing high-dimensional data in lower dimensions to explore its structure.
3. Anomaly Detection
    1. Fraud Detection: Identifying unusual transactions or behaviors in financial data that may indicate fraudulent activity.
    2. Network Security: Detecting unusual network traffic patterns that may signify a security breach or attack.

Q5) What are the main differences between supervised, unsupervised, and semi-supervised learning?

Ans)

1. Type of Data used:
        1. Supervised learning: Uses only labeled data.
        2. Unsupervised learning:Uses a mix of a small amount of labeled data and a large amount of unlabeled data.
        3. Semi supervised learning:Uses both labeled and unlabeled data.
2. Objective:
         1. Supervised learning: Predict or classify based on known labels.
         2. Unsupervised learning:Learn to predict or classify from labeled examples.
         3. Semi supervised learning:Find hidden patterns or groupings without any labels.
3. Training Process:
        1. Supervised learning: Model is trained with explicit feedback from labels.
        2. Unsupervised learning: Model identifies inherent structures without explicit feedback.
        3. Semi supervised learning: Initial training with labeled data followed by refinement with unlabeled data.

Q6. What is train, test, and validation split? Explain the importance of each term

Ans)

1. Training Set:

    Definition: 
        The training set is the subset of the data used to train the machine learning model. It consists of input-output pairs that the model uses to learn the relationships between features and labels.
        
    Importance: 
        1. Model Learning: The model learns the underlying patterns and relationships from the training data.
        2. Parameter Optimization: During training, the model’s parameters are adjusted to minimize the error between predictions and actual outcomes
        
2. Test set:
    
    Definition:
        The test set is a separate and independent subset of the data used to evaluate the final model's performance. It represents new, unseen data that the model has not encountered during training or validation.
    Importance:
        1. Final Evaluation: Provides an unbiased evaluation of the model’s performance after training and tuning.
        2. Performance Metrics: Used to report the final performance metrics (e.g., accuracy, precision, recall) and assess how the model is likely to perform in real-world scenarios.


3. Validation Set
    Definition:
        The validation set is a separate subset of the data used to tune hyperparameters and make decisions about model configurations. It helps in assessing the model's performance during the training process.
        
    Importance:
        1. Hyperparameter Tuning: Used to evaluate different model configurations and select the best hyperparameters.
        2. Model Selection: Helps in selecting the best model version by comparing its performance on the validation data.
        3. Preventing Overfitting: Assists in monitoring the model's performance on data it hasn't seen during training, which helps in detecting and preventing overfitting.

Q7) How can unsupervised learning be used in anomaly detection?

Ans)

Unsupervised learning can be highly effective for anomaly detection, especially when labeled examples of anomalies are rare or unavailable.

How Unsupervised Learning is Used for Anomaly Detection:
        1. Model Training on Normal Data:
                a. Traning: An unsupervised learning model is trained on data that primarily consists of normal or expected patterns, without any explicit labels indicating anomalies.
                b. Learning patterns: The model learns to identify the underlying structure and distribution of the normal data.
        2. Detection of Anomalies:
                a. Score Calculation: Once the model is trained, it can assess new data points by comparing them to the learned normal patterns.
                b. Anomaly Scoring: Data points that deviate significantly from the learned patterns are assigned high anomaly scores, indicating they might be outliers or anomalies.

Common Unsupervised Learning Techniques for Anomaly Detection
1. Clustering Algorithms
2. Dimensionality Reduction
3. Statistical Methods
4. Density-Based Methods

Q8) List down some commonly used supervised learning algorithms and unsupervised algorithms

Ans)

1. Linear Regression:

        Purpose: Predicts a continuous target variable based on one or more input features.
        Example: Predicting house prices based on features like size and location.
        
2. Logistic Regression:

        Purpose: Predicts the probability of a binary outcome based on input features.
        Example: Classifying emails as spam or not spam.
        
3. Decision Trees:

        Purpose: Models decisions and their possible consequences in a tree-like structure.
        Example: Classifying loan applicants as high or low risk based on their financial history.
        
4. Random Forest:

        Purpose: An ensemble method that uses multiple decision trees to improve prediction accuracy and control overfitting.
        Example: Predicting customer churn based on behavioral data
        
5. Support Vector Machines (SVM):

        Purpose: Finds the hyperplane that best separates classes in the feature space.
        Example: Classifying images of handwritten digits.
