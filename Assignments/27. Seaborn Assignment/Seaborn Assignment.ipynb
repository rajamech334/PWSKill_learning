Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans:

1. Scatter Plot

    Use: To visualize the relationship between two continuous variables. It helps in identifying correlations, trends, and potential outliers in the data.

2.	Bar Plot:
        Use: To display the distribution of categorical data and compare the average values of different categories. It is useful for showing summary statistics like mean, median, or count for each category.
        
    
3.	Histogram:


    Use: To represent the distribution of a single continuous variable. It shows the frequency of observations within different intervals (bins) and helps in understanding the data distribution, skewness, and outliers.

4.	Box Plot :


    Use: To summarize the distribution of a dataset using five summary statistics: minimum, first quartile, median, third quartile, and maximum. It helps in identifying outliers and the spread of the data.

5.	Heatmap:

    Use: To represent data in a matrix format where individual values are represented by colors. It is commonly used to visualize correlation matrices, missing values, or any kind of matrix data.


# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = 
"timepoint" and y = "signal" for different events and regions. 

```python
# asn)
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True)
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 
'pclass', y = 'age' and y = 'fare'.

```python
# Ans)
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.boxplot(data=titanic_data, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Age Distribution by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')
sns.boxplot(data=titanic_data, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Fare Distribution by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue 
parameter for the 'cut' column of the diamonds dataset. 

```python
#And)
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = sns.load_dataset("diamonds")
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack')
plt.title('Distribution of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt


flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot("month", "year", "passengers")
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Monthly Passengers Over the Years')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
