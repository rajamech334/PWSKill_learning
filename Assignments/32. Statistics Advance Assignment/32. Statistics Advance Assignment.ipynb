Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with 
an example.

Ans)

The Probability Mass Function (PMF) and Probability Density Function (PDF) are two fundamental concepts in probability theory and statistics, used to describe the probability distributions of discrete and continuous random variables, respectively.

1. Probability Mass Function (PMF):
The Probability Mass Function (PMF) is used for discrete random variables. It provides the probability that a discrete random variable is exactly equal to some value.
 

```python
#   Example PMF Example (Discrete Distribution)
import matplotlib.pyplot as plt
import numpy as np

# Possible outcomes of rolling a fair six-sided die
x = np.arange(1, 7)
# Each outcome has an equal probability of 1/6
pmf = np.ones_like(x) / len(x)

plt.stem(x, pmf, use_line_collection=True)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('PMF of a Fair Six-Sided Die')
plt.ylim(0, 0.2)
plt.show()
```

2. Probability Density Function (PDF):
The Probability Density Function (PDF) is used for continuous random variables. It describes the relative likelihood for a random variable to take on a given value.

```python
#PDF Example (Continuous Distribution)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a range of values from -4 to 4
x = np.linspace(-4, 4, 1000)
# Compute the PDF of the standard normal distribution
pdf = norm.pdf(x, loc=0, scale=1)

plt.plot(x, pdf, label='Standard Normal Distribution')
plt.fill_between(x, pdf, alpha=0.2)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('PDF of the Standard Normal Distribution')
plt.legend()
plt.show()
```

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Ans:
The Cumulative Distribution Function (CDF) of a random variable "X" is a function that gives the probability that "X" will take a value less than or equal to "x". It provides a way to describe the probability distribution of a random variable in terms of cumulative probabilities

```python
#Example of CDF 
# 1. Discrete Case: Fair Six-Sided Die

import numpy as np
import matplotlib.pyplot as plt

# Possible outcomes of rolling a fair six-sided die
x = np.arange(1, 7)
# PMF of each outcome
pmf = np.ones_like(x) / len(x)
# CDF is the cumulative sum of the PMF
cdf = np.cumsum(pmf)

plt.step(x, cdf, where='post', label='CDF')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('CDF of a Fair Six-Sided Die')
plt.ylim(0, 1.1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Continuous Case: Standard Normal Distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a range of values from -4 to 4
x = np.linspace(-4, 4, 1000)
# Compute the CDF of the standard normal distribution
cdf = norm.cdf(x, loc=0, scale=1)

plt.plot(x, cdf, label='Standard Normal CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of the Standard Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

Why CDF is Used:
1. Probability Calculation - The CDF is used to calculate the probability that a random variable falls within a certain range
2. Comparing Distributions - The CDF can be used to compare different probability distributions. It provides a cumulative view that makes it easier to see differences in distributions.
3. Quantile Calculation - The CDF is used to find quantiles (e.g., median, quartiles)
4. Visualizing Distribution - Plotting the CDF provides a way to visualize the distribution of a random variable
5. Statistical Inference - The CDF is fundamental in various statistical methods and hypothesis testing


Q3: What are some examples of situations where the normal distribution might be used as a model? 
Explain how the parameters of the normal distribution relate to the shape of the distribution.

Ans:
The normal distribution, also known as the Gaussian distribution, is widely used in statistics and various fields due to its natural occurrence in many situations and its desirable mathematical properties.
Examples of Situations:
1. Heights of People: The heights of people in a specific population (e.g., adult males in a country) often follow a normal distribution. This is because height is influenced by many small, independent factors (genetics, nutrition, etc.), leading to a bell-shaped distribution

2. Measurement Errors: Measurement errors in scientific experiments or industrial processes often follow a normal distribution. When multiple independent errors contribute to a final measurement, their combined effect tends to be normally distributed.

3. IQ Scores: IQ scores are designed to follow a normal distribution with a mean of 100 and a standard deviation of 15. This standardization allows for comparisons across different populations.

4. Stock Market Returns: Daily returns of stock prices can sometimes be modeled as a normal distribution, especially over short time periods.

Parameters of the Normal Distribution:
The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ).
1. Mean(μ): 
    1. The mean is the central value of the distribution. It determines the location of the peak of the bell curve.
    2. A higher mean shifts the entire distribution to the right, while a lower mean shifts it to the left.
    
2. Standard Deviation (σ):
    1. The standard deviation measures the spread or dispersion of the distribution. It determines the width of the bell curve.
    2. A larger standard deviation results in a wider and flatter bell curve, indicating more variability in the data.
    3. A smaller standard deviation results in a narrower and taller bell curve, indicating less variability

3. Shape of the Distribution
    1. Symmetry: The normal distribution is symmetric about the mean. The left and right sides of the curve are mirror images of each other
    2. Bell-Shaped Curve:The classic bell shape is a hallmark of the normal distribution. Most data points cluster around the mean, and the probabilities taper off equally in both directions as you move further away from the mean.
    3. Inflection Points:The points where the curvature changes direction (from concave to convex) are located at μ−σ and μ+σ. These points indicate one standard deviation away from the mean
    



Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli 
Distribution and Binomial Distribution?

Ans:

The Bernoulli distribution is one of the simplest and most fundamental probability distributions. It models a random experiment that has exactly two possible outcomes: success (with probability p) and failure (with probability 1−p). This distribution is named after the Swiss mathematician Jacob Bernoulli

Example of Bernoulli Distribution:
Consider a simple experiment of flipping a fair coin:
    1. the probability of getting heads (success) is p=0.5.
    2. The probability of getting tails (failure) is 1−p=0.5.
    
Bernoulli Distribution vs. Binomial Distribution:
While the Bernoulli distribution models a single trial, the Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials
Key differences are:
1. Number of Trials:
    1. Bernoulli Distribution: Models a single trial (n = 1).
    2. Binomial Distribution: Models multiple trials (n ≥ 1).
2. Random Variable:
    1. Bernoulli Distribution: The random variable "X" takes on two possible values, 0 or 1.
    2. Binomial Distribution: The random variable "X" represents the number of successes in n trials and can take on any integer value from 0 to n.


Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset 
is normally distributed, what is the probability that a randomly selected observation will be greater 
than 60? Use the appropriate formula and show your calculations.

Ans:

Step-1: 
        Convert the value to a z-score:
        The formula to convert a value X to a z-score Z is: (X−μ)/σ
        So the Z value will be Z= (60-50)/10
        Z = 1
        
Step-2: 
        Find the probability associated with the z-score
        We need to find the probability that "Z" is greater than 1. This is denoted as P(Z>1).
        The cumulative distribution function of the standard normal distribution gives the probability that Z is less than or equal   to a particular value. We denote this as Φ(z), where Φ(1) is the CDF value at Z=1.
        
step-3:
        Use standard normal distribution tables or a calculator - The CDF value for z=1 is 0.8413.

step-4:
        Calculate the probability that Z is greater than 1: P(Z>1)=1−0.8413=0.1587
        
 

```python
# Sample program for reference

from scipy.stats import norm

# Parameters
mu = 50
sigma = 10
x = 60

# Calculate the z-score
z = (x - mu) / sigma

# Calculate the probability P(X > 60)
p = 1 - norm.cdf(z)

print(f"The probability that a randomly selected observation is greater than {x} is approximately {p:.4f}.")
```

Q7: Explain uniform Distribution with an example.

Ans:

The uniform distribution is a type of probability distribution in which all outcomes are equally likely. There are two main types of uniform distributions: discrete uniform distribution and continuous uniform distribution.



```python
'''
Example: Consider the scenario where a bus arrives at a bus stop at a random time between 0 and 10 minutes. 
Any time within this interval is equally likely
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Discrete Uniform Distribution (rolling a fair six-sided die)
x_discrete = np.arange(1, 7)
pmf_discrete = np.ones_like(x_discrete) / len(x_discrete)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem(x_discrete, pmf_discrete, use_line_collection=True)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution (Fair Die)')
plt.xticks(x_discrete)

# Continuous Uniform Distribution (bus arrival time between 0 and 10 minutes)
a, b = 0, 10
x_continuous = np.linspace(a, b, 1000)
pdf_continuous = uniform.pdf(x_continuous, loc=a, scale=b-a)

plt.subplot(1, 2, 2)
plt.plot(x_continuous, pdf_continuous, label='PDF')
plt.fill_between(x_continuous, pdf_continuous, alpha=0.2)
plt.xlabel('Time (minutes)')
plt.ylabel('Density')
plt.title('Continuous Uniform Distribution (Bus Arrival Time)')
plt.legend()

plt.tight_layout()
plt.show()
```

Q8: What is the z score? State the importance of the z score.

Ans:
A z-score, also known as a standard score, indicates how many standard deviations an element is from the mean of the distribution. It is a way to describe the position of a raw score in terms of its distance from the mean, measured in standard deviations

Importance of the Z-Score:
1. Standardization:
    Z-scores standardize different distributions to a common scale, allowing for comparisons across different datasets. By converting raw scores to z-scores, any normal distribution can be transformed into the standard normal distribution (mean of 0 and standard deviation of 1).
    
2. Identifying Outliers:
    Z-scores help identify outliers in the data. A z-score beyond ±2 or ±3 is often considered an outlier, indicating that the data point is significantly different from the mean.
    
3. Probability Calculations:
    Z-scores are essential in calculating probabilities in a normal distribution. By converting a raw score to a z-score, one can use standard normal distribution tables or statistical software to find the probability of a score occurring below, above, or between certain values.
    
4. Comparison Across Different Scales:
    When different datasets have different means and standard deviations, z-scores enable comparison by placing scores on a common scale. For example, comparing test scores from different exams becomes possible through z-scores.

5. Hypothesis Testing:
    Z-scores play a crucial role in hypothesis testing, especially in z-tests. They help determine whether to reject the null hypothesis by comparing the z-score to critical values.
    
6. Normalization:
    In machine learning and data preprocessing, z-scores are used to normalize data, ensuring that features have a mean of 0 and a standard deviation of 1.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans:

The Central Limit Theorem is a fundamental theorem in statistics that describes the distribution of the sample mean of a large number of independent, identically distributed random variables.

When a sufficiently large sample size n is drawn from any population with a finite mean μ and finite variance the distribution of the sample mean will be approximately normally distributed, regardless of the shape of the population distribution. The mean of this sampling distribution will be equal to the population mean and its standard deviation.

Significance of the Central Limit Theorem:

1. Foundation for Inferential Statistics - 
    The CLT is the cornerstone of inferential statistics. It allows statisticians to make inferences about population parameters based on sample statistics, even when the population distribution is not normal.
    
2. Justification for the Use of Normal Distribution:
    Because the CLT states that the distribution of the sample mean approximates a normal distribution for large sample sizes, many statistical methods assume normality for the sake of simplicity and tractability.

3. Facilitates Hypothesis Testing and Confidence Intervals:
    The approximation to normality allows for the use of z-tests and t-tests for hypothesis testing. It also aids in constructing confidence intervals for population parameters, as the sample mean can be treated as normally distributed
    
4. Application in Quality Control:
    In industrial and quality control processes, the CLT is used to monitor the consistency of processes. By sampling a small number of items from a production line, the average quality of the items can be assessed.

5. Simplifies Analysis of Large Datasets:
    For large datasets, the CLT helps in understanding the behavior of sample means, making it easier to analyze and interpret the data.


Q10: State the assumptions of the Central Limit Theorem.

Ans:

The Central Limit Theorem (CLT) is a fundamental theorem in statistics, but it relies on certain assumptions to hold true. These assumptions are crucial for the theorem to accurately describe the behavior of sample means from a population.

Assumptions of the Central Limit Theorem:

1. Independence:
    The samples drawn from the population must be independent of each other. This means that the outcome of one sample should not influence the outcomes of other samples.

2. Identically Distributed:
    The samples should be drawn from the same population and have the same probability distribution. This assumption ensures that each sample follows the same underlying distribution with the same parameters (mean and variance).
    
3. Finite Variance:
    The population from which the samples are drawn should have a finite variance. If the variance is infinite or undefined, the sample means may not converge to a normal distribution as predicted by the CLT.

4. Large Sample Size:
    For the CLT to provide accurate approximations, the sample size "n" should be sufficiently large. While there is no strict cutoff for what constitutes a "large" sample size, a common guideline is "n≥30". Larger sample sizes tend to provide better approximations to the normal distribution
