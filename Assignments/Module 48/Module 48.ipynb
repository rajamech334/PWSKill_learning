```python
"""Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation 
of 5 using Python. Interpret the results.

Ans)
Assuming sample data n = 30"""

import scipy.stats as stats
import math

mean = 50
std_dev = 5
sample_size = 30
confidence_level = 0.95
# Z-score for 95% confidence
z_score = stats.norm.ppf((1 + confidence_level) / 2)
# Calculate the margin of error
margin_of_error = z_score * (std_dev / math.sqrt(sample_size))
# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
confidence_interval

```

```python
"""Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag 
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% 
brown. Use Python to perform the test with a significance level of 0.05."""

#Ans)

import scipy.stats as stats

# Assuming observed frequesncies are 
observed_frequencies = [18, 22, 20, 8, 12, 20]

# Expected frequencies based on the expected distribution
total_mms = sum(observed_frequencies)
expected_frequencies = [total_mms * 0.20, total_mms * 0.20, total_mms * 0.20, total_mms * 0.10, total_mms * 0.10, total_mms * 0.20]

# Perform the chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)

chi2_stat, p_value

```

```python
"""Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals 
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the 
population who smoke."""

#Ans)

import math

# Given values
n = 500
x = 60
confidence_level = 0.95

# Calculating the sample proportion
p_hat = x / n

# Z-score for 95% confidence
z_score = 1.96

# Calculating the margin of error
margin_of_error = z_score * math.sqrt((p_hat * (1 - p_hat)) / n)

# Calculating the confidence interval
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)
confidence_interval
```

```python
"""Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation 
of 12 using Python. Interpret the results."""

#Ans)
import scipy.stats as stats
import math

# Given values
mean = 75
std_dev = 12
sample_size = 30
confidence_level = 0.90

# Z-score for 90% confidence
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_score * (std_dev / math.sqrt(sample_size))

# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
confidence_interval
```

```python
"""Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the 
area corresponding to a chi-square statistic of 15."""
#Ans)
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Define the degrees of freedom and the chi-square statistic
df = 10
chi2_stat = 15

# Generate a range of values for the x-axis
x = np.linspace(0, 30, 1000)

# Calculate the chi-square distribution PDF
pdf = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Chi-square distribution with {df} degrees of freedom')
plt.fill_between(x, pdf, where=(x <= chi2_stat), color='skyblue', alpha=0.5)
plt.axvline(x=chi2_stat, color='red', linestyle='--', label=f'Chi-square statistic = {chi2_stat}')

# Label the axes and the plot
plt.title('Chi-square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-square value')
plt.ylabel('Probability density')
plt.legend()

# Show the plot
plt.show()
```

```python
"""Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who 
prefer Coke."""

#Ans)

import math

# Input Values
n = 1000
x = 520
confidence_level = 0.99

# Calculate the sample proportion
p_hat = x / n

# Z-score for 99% confidence
z_score = 2.576

# Calculate the margin of error
margin_of_error = z_score * math.sqrt((p_hat * (1 - p_hat)) / n)

# Calculate the confidence interval
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)
confidence_interval
```

```python
"""Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the 
expected frequencies of a fair coin. Use a significance level of 0.05."""

#Ans)

import scipy.stats as stats

# Observed frequencies
observed_frequencies = [45, 55]

# Expected frequencies for a fair coin
expected_frequencies = [50, 50]

# Perform the chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)

chi2_stat, p_value
```

```python
"""Q9. A study was conducted to determine if there is an association between smoking status (smoker or 
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. 
Conduct a chi-square test for independence to determine if there is a significant association between 
smoking status and lung cancer diagnosis."""

#Ans)

import scipy.stats as stats
# Define the contingency table
contingency_table = [
    [60, 140],
    [30, 170]
]
# Perform the chi-square test for independence
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard 
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. 
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a 
significance level of 0.05.

Ans)
Conducting Hypothesis testing

Step-1: Stating the Hypothesis
    1. Null Hypothesis - The population mean is equal to 70
    2. Alternative Hypothesis - The population mean is not equal to 70

Step-2: Set the significance level 
    The significance level (α) is given as 0.05.
    
Step-3: Calculate the Test Statistic
        t = (72-70)/10√30 = 1.091
Step-4: Determine critical value
    For a two-tailed test with α=0.05 and degrees of freedom 𝑑𝑓=𝑛−1=30−1=29
    The critical t-value for 𝑑𝑓=29 df=29 at 𝛼=0.05 α=0.05 (two-tailed) is approximately ±2.045.
    
Step-5: Compare the Test Statistic to the Critical Value
    1. If the test statistic is within the range of −2.045 to 2.045, we fail to reject the null hypothesis.
    2. If the test statistic is outside this range, we reject the null hypothesis
Result:

Since the test statistic does not fall outside the critical range, we fail to reject the null hypothesis. There is not enough evidence to conclude that the population mean is significantly different from 70 at the 0.05 significance level
