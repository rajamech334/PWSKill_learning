Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans)

Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating different types of charts, plots, and graphs. It is used for visualizing data in a clear and concise manner, which helps in identifying patterns, trends, and outliers in the data. Matplotlib can be used for a variety of tasks, such as data exploration, data analysis, and data communication.

The pyplot module of Matplotlib provides a simple interface for creating different types of plots, including:

1. Line plot - used to display data points as a series of connected lines.
2. Scatter plot - used to display the relationship between two variables as a collection of points on a two-dimensional plane.
3. Bar plot - used to compare different categories of data using rectangular bars.
4. Histogram - used to display the distribution of a dataset using a series of contiguous rectangular bars or bins.
5. Pie chart - used to display the composition of a dataset by dividing it into slices that represent different proportions of the whole.
There are many other types of plots that can be created using Matplotlib, such as area plots, box plots, violin plots, and more.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.


Ans)

A scatter plot is a type of plot that displays the relationship between two variables by representing each observation as a point on a two-dimensional plane. It is useful for identifying patterns or trends in the data, as well as for detecting outliers or unusual observations.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans)

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows us to arrange plots in a grid of rows and columns, and then specify which plot to draw on using its position in the grid.

```python
import numpy as np
import matplotlib.pyplot as plt

# data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create subplots
fig, axs = plt.subplots(2, 2)

# plot for line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# plot for line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# plot for line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# plot for line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# add a common x-label and y-label for all subplots
fig.text(0.5, 0.04, 'X', ha='center')
fig.text(0.04, 0.5, 'Y', va='center', rotation='vertical')

# adjust spacing between subplots
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.4, hspace=0.4)

# display the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])


Ans)
A bar plot is a type of plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent. It is used to compare and visualize the differences in values across categories.

To create a bar plot and a horizontal bar plot using the given data, we can use the bar() and barh() functions respectively from the Pyplot module of Matplotlib.


```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating a bar plot
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Creating a horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans)

A box plot is a graphical representation of the distribution of numerical data through their quartiles. It is used to identify the median, the interquartile range, and the outliers of a dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

# generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# create figure and axes
fig, ax = plt.subplots()

# create box plot
box_plot_data = [box1, box2]
ax.boxplot(box_plot_data)

# set title and axis labels
ax.set_title('Box Plot')
ax.set_xlabel('Data')
ax.set_ylabel('Value')

# show plot
plt.show()
```
